```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix

X, y = datasets.make_classification(n_samples=500, n_features=2, n_informative=2, n_redundant=0, random_state=42)
y = np.where(y == 0, -1, 1)  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

kernels = ['linear', 'rbf', 'poly']
results = {}

for kernel in kernels:
    clf = SVC(kernel=kernel, degree=3 if kernel == 'poly' else None, random_state=42)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    results[kernel] = {
        'model': clf,
        'accuracy': acc,
        'confusion_matrix': cm
    }

def plot_decision_boundary(model, X, y, title):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', marker='o')
    plt.title(title)
    plt.show()

for kernel in kernels:
    model = results[kernel]['model']
    plot_decision_boundary(model, X_test, y_test, f'{kernel.capitalize()} Kernel Decision Boundary')

for kernel in kernels:
    print(f"Kernel: {kernel.capitalize()}")
    print(f"Accuracy -> {results[kernel]['accuracy']:.2f}")
    print(f"confusion matrix ->\n{results[kernel]['confusion_matrix']}\n")

#sources: pytorch
```
