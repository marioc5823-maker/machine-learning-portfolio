Monty Hall Problem

```python
import sys
import pandas as pd
import numpy as np
import random

while True:
    print("Choose from one of the doors:(1,2,3) \n")
    doorValue = int(input())
    if doorValue in {1,2,3}:
        break

print("You chose: ", doorValue)

myArray = [0,0,0]
winningDoor = random.randint(0,len(myArray) - 1)
myArray[winningDoor] = '1'

print(myArray)

chosenDoor = myArray[doorValue - 1]

if chosenDoor == 0: 
    print("you lose")
else:
    print("you win")

falseDoor = next(i for i in range(3) if i != winningDoor and i != doorValue - 1)

print("Considering door", falseDoor + 1, "is empty")
choice = input("Would you like to switch doors? (Y/N) ").strip().lower()

if choice == 'y':
    doorValue = next(i for i in range(3) if i != doorValue - 1 and i != falseDoor) + 1
    print("You switched doors!")
elif choice == 'n':
    print("You kept your original choice!")
else:
    print("Invalid input! Please enter 'y' or 'n'.")

if myArray[doorValue - 1] == '1':
    print("Final result: You win!")
else:
    print("Final result: You lose!")
```

```python
import random
import pandas as pd
import sys

def monty_hall(change_choice: bool) -> bool:
    doors = [0, 0, 0]
    prize_door = random.randint(0, 2)
    doors[prize_door] = 1

    player_pick = random.randint(0, 2)

    available_doors = [i for i in range(3) if i != player_pick and doors[i] != 1]
    revealed_door = random.choice(available_doors)

    if change_choice:
        player_pick = next(i for i in range(3) if i != player_pick and i != revealed_door)

    return doors[player_pick] == 1

def run_experiment(trials: int) -> dict:
    outcomes = {'change': 0, 'stay': 0}
    
    for _ in range(trials):
        if monty_hall(change_choice=True):
            outcomes['change'] += 1
        if monty_hall(change_choice=False):
            outcomes['stay'] += 1

    return outcomes

total_trials = 10000
results = run_experiment(total_trials)

change_win_rate = (results['change'] / total_trials) * 100
stay_win_rate = (results['stay'] / total_trials) * 100

print("Assuming each door was randomly assigned, we concluded the following:")
print("After ", total_trials," Trials,")
print(f"Changing Choice Wins: {results['change']} ({change_win_rate:.2f}%)")
print(f"Staying Choice Wins: {results['stay']} ({stay_win_rate:.2f}%)")
```
