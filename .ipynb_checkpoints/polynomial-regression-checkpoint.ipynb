```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures

mtcars_data = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv")[['mpg', 'wt']]

plt.scatter(mtcars_data['wt'], mtcars_data['mpg'], color='blue')
plt.title('wt vs mpg')
plt.xlabel('wt:')
plt.ylabel('mpg:')
plt.grid()
plt.show()

X = mtcars_data[['wt']]
y = mtcars_data['mpg']
linear_model = LinearRegression().fit(X, y)
print(f'Linear Intercept: {linear_model.intercept_}, Coefficient: {linear_model.coef_[0]}, MSE: {mean_squared_error(y, linear_model.predict(X))}')

plt.scatter(mtcars_data['wt'], mtcars_data['mpg'], color='blue')
plt.plot(X, linear_model.predict(X), color='red')
plt.title('LR: mpg vs wt')
plt.show()

poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)
quadratic_model = LinearRegression().fit(X_poly, y)
print(f'Quadratic Intercept: {quadratic_model.intercept_}, Coefficients: {quadratic_model.coef_[1:]}, MSE: {mean_squared_error(y, quadratic_model.predict(X_poly))}')

plt.scatter(mtcars_data['wt'], mtcars_data['mpg'], color='blue')
plt.scatter(mtcars_data['wt'], quadratic_model.predict(X_poly), color='red', alpha=0.5)
plt.title('Quadratic Regression: Weight vs. MPG')
plt.show()
```
